```python
# Define basic Python datatypes and variables

# Integer
age = 25
count = 100

# Float
price = 19.99
temperature = 98.6

# String
name = "Alice"
city = "New York"
message = "Hello, World!"

# Boolean
is_active = True
is_student = False

# List
fruits = ["apple", "banana", "orange", "grape"]
numbers = [1, 2, 3, 4, 5]

# Tuple
coordinates = (10, 20)
rgb_color = (255, 128, 0)

# Dictionary
person = {
    "name": "Bob",
    "age": 30,
    "email": "bob@example.com"
}
settings = {"theme": "dark", "notifications": True}

# Set
unique_colors = {"red", "blue", "green"}
unique_numbers = {1, 2, 3, 4, 5}

# None
result = None

# Display variables
print(f"Name: {name}, Age: {age}")
print(f"Fruits: {fruits}")
print(f"Person: {person}")
print(f"Coordinates: {coordinates}")
```

```python
# Control flow with if statements

# Simple if statement
if age >= 18:
    print(f"{name} is an adult (age: {age})")

# if-else statement
if is_student:
    print("Currently a student")
else:
    print("Not a student")

# if-elif-else statement
if temperature > 100:
    print("Hot")
elif temperature > 32:
    print("Warm")
else:
    print("Cold")

# Check if item exists in list
if "banana" in fruits:
    print(f"Banana is in the fruits list")

# Check dictionary value
if person["age"] > age:
    print(f"{person['name']} is older than {name}")

# Check if result is None
if result is None:
    print("Result is not set yet")
```

```python
# For loop with list
print("Fruits:")
for fruit in fruits:
    print(f"  - {fruit}")

# For loop with range
print("\nNumbers squared:")
for num in numbers:
    print(f"  {num}^2 = {num**2}")

# For loop with enumerate
print("\nFruits with index:")
for index, fruit in enumerate(fruits):
    print(f"  {index}: {fruit}")

# For loop with dictionary
print("\nPerson details:")
for key, value in person.items():
    print(f"  {key}: {value}")

# While loop
print("\nCountdown from 5:")
i = 5
while i > 0:
    print(f"  {i}")
    i -= 1

# Loop with break
print("\nFind first fruit with 'a':")
for fruit in fruits:
    if 'a' in fruit:
        print(f"  Found: {fruit}")
        break

# Loop with continue
print("\nNumbers (skip even):")
for num in numbers:
    if num % 2 == 0:
        continue
    print(f"  {num}")
```
