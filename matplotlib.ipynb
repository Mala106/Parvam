```python
import numpy as np

import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Line Chart
axes[0, 0].plot(x, y, color='blue', linewidth=2)
axes[0, 0].set_title('Line Chart - Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].grid(True)

# 2. Bar Chart
axes[0, 1].bar(categories, values, color='green', alpha=0.7)
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_ylabel('Values')

# 3. Scatter Plot
axes[1, 0].scatter(x, y, color='red', alpha=0.5, s=50)
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')

# 4. Pie Chart
axes[1, 1].pie(values, labels=categories, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow', 'purple'])
axes[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
```

```python
# Display the figure with all subplots
print("Displaying 4 different chart types:")
print("- Line Chart: Sine wave pattern")
print("- Bar Chart: Sample category values")
print("- Scatter Plot: Scattered sine wave points")
print("- Pie Chart: Distribution of values across categories")
print("\nThe plots are displayed above.")
```
